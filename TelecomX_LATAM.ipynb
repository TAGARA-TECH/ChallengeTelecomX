#📌 Extracción

```python
import pandas as pd
import requests

# URL del archivo JSON crudo
url = 'https://raw.githubusercontent.com/TAGARA-TECH/ChallengeTelecomX/main/TelecomX_Data.json'

# Obtener los datos desde la "API"
response = requests.get(url)
data_json = response.json()

# Convertir a DataFrame
df = pd.DataFrame(data_json)

# Mostrar las primeras filas
df.head()
```

```python
# Vista general del DataFrame
print("Dimensiones del dataset:", df.shape)

# Columnas y tipos de datos
df.info()

# Tipos de datos por separado
print("\nTipos de datos:")
print(df.dtypes)

# Primeras filas para exploración visual
df.head()
```

```python
# Estadísticas de variables numéricas
df.describe()

# Ver valores únicos por columna (para identificar categóricas)

for col in df.columns:
    try:
        unique_vals = df[col].nunique()
        print(f"{col}: {unique_vals} valores únicos")
    except TypeError as e:
        print(f"⚠️ Error en columna '{col}': {e}")
        print(f"🔍 Primeros 2 valores no nulos en '{col}':")
        print(df[col].dropna().head(2).tolist())
```

 el dataset tiene columnas anidadas, es decir, varias columnas contienen diccionarios dentro de cada fila. Para poder analizarlas correctamente en pandas, hay que "aplanarlas" (flatten), es decir, convertir esas claves internas en columnas separadas.

```python
# Expandir la columna 'customer'
customer_df = df['customer'].apply(pd.Series)
df = df.drop('customer', axis=1).join(customer_df)

# Expandir la columna 'phone'
phone_df = df['phone'].apply(pd.Series)
df = df.drop('phone', axis=1).join(phone_df)

# Expandir la columna 'internet'
internet_df = df['internet'].apply(pd.Series)
df = df.drop('internet', axis=1).join(internet_df)

# Expandir la columna 'account'
account_df = df['account'].apply(pd.Series)

# Extraer también los valores anidados en 'Charges'
charges_df = account_df['Charges'].apply(pd.Series)
account_df = account_df.drop('Charges', axis=1).join(charges_df)

# Unir todo al DataFrame original
df = df.drop('account', axis=1).join(account_df)
```

La fase de Extracción se ha completado con éxito. Se ha accedido a los datos del cliente desde la "API" (un archivo JSON alojado en GitHub) utilizando la librería requests de Python, y se ha cargado correctamente en un DataFrame de Pandas.

3. Hallazgos Preliminares y Estructura de Datos
Un análisis inicial del DataFrame ha revelado una estructura particular en los datos:

Las columnas customer, phone, internet y account son de tipo object, lo que indica que contienen estructuras de datos anidadas, específicamente diccionarios. Estas columnas encapsulan información valiosa sobre las características del cliente, sus servicios telefónicos, servicios de internet y detalles de la cuenta.

La columna Churn, que es nuestra variable objetivo (indica si un cliente se fue o no), presenta tres valores únicos: 'No', 'Yes', y una cadena vacía (''). La presencia de cadenas vacías en esta columna es un problema de calidad de datos que debe ser abordado antes de cualquier análisis predictivo.

Para avanzar en el desafío y preparar los datos para un análisis más profundo, los siguientes pasos son cruciales:

Aplanamiento de Columnas Anidadas: La tarea más inmediata es desanidar las columnas customer, phone, internet y account. Cada clave dentro de estos diccionarios anidados debe convertirse en una columna separada en el DataFrame principal. Esto es fundamental para hacer que la información sea accesible y utilizable para el análisis.

Las opciones incluyen la eliminación de estas filas si su cantidad es insignificante o la imputación si es apropiado, aunque para una variable objetivo binaria, la eliminación suele ser la opción más segura si los datos son realmente ausentes.
